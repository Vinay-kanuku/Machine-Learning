# SVD

```python
import pandas as pd 
import numpy as np 
from PIL import Image
```

```python
image = Image.open("DataSets/bt.jpg")
image = image.convert("L")
# image = image.resize((200, 200))
image_as_arrya = np.array(image)
image
```

```python
u, d, vt = np.linalg.svd(image_as_arrya)   #SVD
```

```python

k = 150
u_k = u[:, :k]
d_k = np.diag(d[:k])
vt_k = vt[:k, :]
```

```python
approax_img = np.dot(u_k, np.dot(d_k, vt_k))
approax_img.size
```

```python
original_size = image_as_arrya.size
reconst_size = u_k.size + d_k.size + vt_k.size
compresion_ratio = original_size / reconst_size
compresion_ratio
```
